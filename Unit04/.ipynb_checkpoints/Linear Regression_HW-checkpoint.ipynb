## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
X = x.reshape(-1, 1)

#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

regr = LinearRegression()

regr.fit(x_train, y_train)

y_train_pred = regr.predict(x_train)

plt.scatter(x_train, y_train, color='blue', label='Train data')
plt.plot(x_train, y_train_pred, color='red', label='Linear regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Train Data Line')
plt.legend()
plt.show()
```

```python
mse = np.mean((y_train - y_train_pred) ** 2)
print("Train MSE: " + str(mse))
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
# input data X= 2.44897959
x_value = np.array([[2.44897959]])
y_pred_value = regr.predict(x_value)
print("Predicted value : " + str(y_pred_value))
```

2. 並計算出測試階段的MSE

```python
y_test_pred = regr.predict(x_test)
mse_test =  np.mean((y_test - y_test_pred) ** 2)
print("Test MSE: " + str(mse_test))
```

3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_test.PNG)

```python
title = "f(x)=1.6x+0.2, mse:"+str(round(mse_test,2))
plt.title(title)
plt.xlabel('x')
plt.ylabel('y')

plt.scatter(x_test, y_test, alpha=0.8, label='Ground Truth')
plt.scatter(x_test, y_test_pred, color='red', alpha=0.8, label='Predict')
plt.legend()
plt.show()
```
